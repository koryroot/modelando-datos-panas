```python
#Cambiando los nombres de las columnas

df1 = df1.rename(columns={
    'Your Sex?': 'sexo',
    'Your Matric (grade 12) Average/ GPA (in %)': 'promedio',
    'What year were you in last year (2023) ?': 'last_year',
    'What faculty does your degree fall under?': 'Asignatura',
    'Your 2023 academic year average/GPA in % (Ignore if you are 2024 1st year student)': 'current_year',
    'Your Accommodation Status Last Year (2023)': 'Tipo_escuela',
    'Monthly Allowance in 2023': 'Mesada',
    'Were you on scholarship/bursary in 2023?': 'estudiante_becado',
    'Additional amount of studying (in hrs) per week': 'Horas_semanales_estudios',
    'How often do you go out partying/socialising during the week? ': 'Party_nocturno',
    'On a night out, how many alcoholic drinks do you consume?': 'Cantidad_de_shop',
    'How many classes do you miss per week due to alcohol reasons, (i.e: being hungover or too tired?)': 'Clases_perdidas_por_resaca',
    'How many modules have you failed thus far into your studies?': 'Asignaturas_quemadas',
    'Are you currently in a romantic relationship?': 'Novio/a',
    'Do your parents approve alcohol consumption?': 'Padres_aprueban_alcohol',
    'How strong is your relationship with your parent/s?': 'Buena_relacion_con_tus_padres',
})
```

```python
df1['sexo'].fillna(df1['sexo'].mode()[0], inplace=True)
df1['promedio'].fillna(df1['promedio'].mode()[0], inplace=True) #puedo rellenar con ceros esta opcion
df1['last_year'].fillna(df1['last_year'].mode()[0], inplace=True)
df1['Asignatura'].fillna(df1['Asignatura'].mode()[0], inplace=True)
df1['current_year'].fillna(df1['current_year'].mode()[0], inplace=True) #puedo rellenar con ceros esta opcion
df1['Tipo_escuela'].fillna(df1['Tipo_escuela'].mode()[0], inplace=True)
df1['Mesada'].fillna(df1['Mesada'].mode()[0], inplace=True)
df1['estudiante_becado'].fillna(df1['estudiante_becado'].mode()[0], inplace=True)
df1['Horas_semanales_estudios'].fillna(df1['Horas_semanales_estudios'].mode()[0], inplace=True)
df1['Party_nocturno'].fillna(df1['Party_nocturno'].mode()[0], inplace=True)
df1['Cantidad_de_shop'].fillna(df1['Cantidad_de_shop'].mode()[0], inplace=True)
df1['Clases_perdidas_por_resaca'].fillna(df1['Clases_perdidas_por_resaca'].mode()[0], inplace=True)
df1['Asignaturas_quemadas'].fillna(df1['Asignaturas_quemadas'].mode()[0], inplace=True)
df1['Novio/a'].fillna(df1['Novio/a'].mode()[0], inplace=True)
df1['Padres_aprueban_alcohol'].fillna(df1['Padres_aprueban_alcohol'].mode()[0], inplace=True)
df1['Buena_relacion_con_tus_padres'].fillna(df1['Buena_relacion_con_tus_padres'].mode()[0], inplace=True)

df1.tail()
```

```python
for column in df1.select_dtypes(include='number'):
    df_temp = df1[column].to_frame()

    # Generando un grafico por columna
    df_temp.boxplot()
    plt.title(f"Boxplot para la columna: {column}")
    plt.show()
```

```python
for column in df1.select_dtypes(include='number'):
    df1.boxplot(column=column)
    plt.title(f"Boxplot para la columna: {column}")

    # Calculando los IQR
    Q1 = df1[column].quantile(0.25)
    Q3 = df1[column].quantile(0.75)
    IQR = Q3 - Q1

    # Identificando los outliers
    upper_bound = Q3 + 1.5 * IQR
    lower_bound = Q1 - 1.5 * IQR
    outliers = df1[(df1[column] > upper_bound) | (df1[column] < lower_bound)]

    # Marcando los outliers en el boxplot
    plt.scatter(outliers.index, outliers[column], color='red')

    plt.show()
```
